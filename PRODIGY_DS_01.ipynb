Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

Dataset:Student data set
Source:Kaggle

Description:
Categorical-Gender
           >The gender breakdown using Histogram is done.Insights into 'Male'and 'Female' counts,disclosing the population's gender.
Continuous-Age demographics
            >Age distribution with a detailed histogram is created.The frequency of different age groups points out a clear picture of the population's age demographics.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model,metrics
```

```python
#load domain dataset
import pandas as pd
df=pd.read_csv("student_data.csv")
df
```

```python
#continuous data
plt.figure(figsize=(8,6))
plt.hist(df['age'],bins=20,color='red',edgecolor='yellow')

plt.title('Histogram of Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Categorical data
plt.figure(figsize=(8,6))
plt.hist(df['sex'],bins=20,color='green',edgecolor='yellow')

plt.title('Histogram of Sex distribution')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.show()
```
